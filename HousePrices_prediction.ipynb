```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

Loading Dataset

```python
df=pd.DataFrame({
    'area':[1000,2000,3000,4000,5000,6000,7000],
    'bedrooms':[3,4,2,0,3,4,2],
    'price':[58,65,77,83,96,110,125]
})
```

Visualize Data

```python
plt.scatter(df['area'], df['price'], color='blue',marker='*')
plt.xlabel("Area (sqft)")
plt.ylabel("Price (Lakhs)")
plt.title("Area vs Price")
plt.show()
```

Train the model

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(df[['area','bedrooms']],df['price'])
```

Prediction Price

```python
area_value = 3300
bedrooms_value = 3
predicted_price = model.predict([[area_value,bedrooms_value]])
print(f"Predicted price for {area_value} sqft: ₹{predicted_price[0]:.2f} Lakhs")
```

Plot Regression Line

```python
plt.scatter(df['area'], df['price'], color='blue')
plt.plot(df['area'], model.predict(df[['area','bedrooms']]), color='red')
plt.xlabel("Area")
plt.ylabel("Price")
plt.title("Regression Line: Area vs Price")
plt.show()
```

```python
model.score(df[['area','bedrooms']],df['price'])
```
